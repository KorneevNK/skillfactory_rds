```python
pivot_1=df.pivot_table(index=['Club'],columns=['Position'],margins=True,values=['Name'],fill_value=0,aggfunc='count').reset_index()
sumka=0
v=0
a=[]
for i in range(0,650):
    sumka+=pivot_1.loc[i]['Name']['GK']
    a.append(i)
    v=sumka/len(a)
print(round(v,3))
```
