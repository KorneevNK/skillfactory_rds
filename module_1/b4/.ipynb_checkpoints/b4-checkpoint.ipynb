```python
log = pd.read_csv('files/log.xls', header=None)
log.columns=['user_id','time','bet','win']
def change_id(x):
    if 'error' in x: x=''
    else: x=x.replace('Запись пользователя № - ','')
    return x
log = log[-log.user_id.str.contains('error', na=False)]
log.user_id = log.user_id.apply(change_id)
log.time = log.time.str.replace('[','')
log
```
